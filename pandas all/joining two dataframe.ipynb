```python
df_1 = pd.DataFrame({'id':[1,2,3,4,5],'Class':[6,7,8,9,10]})
df_1
```

```python
df_2 = pd.DataFrame({'id':[1,2,3,4,6],'Name':['a','b','c','d','e']})
df_2
```

```python
pd.merge(df_2,df_1,how = 'outer')# union and intersection
```

```python
pd.merge(df_2,df_1,how = 'outer',indicator=True)
```

```python
pd.merge(df_2,df_1,how = 'outer',right_index=True,left_index=True,suffixes=['_Rifat','_Fariha'])#removing thme using on
```

```python
pd.merge(df_2,df_1,how = 'outer',sort=True)
```
