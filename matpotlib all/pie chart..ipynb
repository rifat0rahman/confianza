```python
from matplotlib import pyplot as plt
from matplotlib import style as st
import numpy as np
import random
import pandas as pd
import  matplotlib.image as img
```

```python
plt.pie([1])
plt.show()
```

```python
Classes = ['FIVE','SIX','SEVEN','EIGHT','NINE','TEN']
male_Students = [40,56,45,60,42,50]

plt.pie(male_Students,labels=Classes,explode=[0.01,0.01,0.1,0.01,0.01,0.01],autopct='%0.2f%%',
       shadow = True)
plt.legend(loc =[1.1,0.3]) #here can be used multiple for placing
plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1])
plt.subplot(2,2,2)
plt.pie([1,2])
plt.subplot(2,2,3)
plt.pie([1,2,3])
plt.subplot(2,2,4)
plt.pie([1,2,3,4])
plt.show()
```
