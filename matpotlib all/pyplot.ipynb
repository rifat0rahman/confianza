```python
from matplotlib import pyplot as plt
from matplotlib import style as st
import pandas as pd
import numpy as np
import random
```

## plot or অজিবরেখা

```python
students =[23,43,45,23,67,87,54,32,21,32]
classes = np.arange(10)
plt.title('this is just a demo')
plt.xlabel('students')
plt.ylabel('classes')
plt.plot(students,classes)
plt.show()
```

```python
xxis = np.arange(1,11)
yxis = np.random.randint(40,100,10)
```

```python
st.use('ggplot')
plt.plot(xxis,yxis,color = 'b',marker = 'o',markeredgecolor='k',linewidth = 3,
         markerfacecolor = 'k',ms  = 6,label = 'par')
plt.title('Just a moment with my life',fontsize = 20)
plt.xlabel('Classes',fontsize = 20)
plt.ylabel('Students',fontsize = 20)
plt.legend()
plt.grid(color = 'w')
plt.show()
```
