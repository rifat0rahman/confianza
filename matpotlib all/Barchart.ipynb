```python
from matplotlib import pyplot as plt
from matplotlib import style as st
import numpy as np
import random
import pandas as pd
import  matplotlib.image as img
```

```python
Platform = ['facebook','instragram','pubg','free fire','tiktok','youtube']
Users = [56,87,98,45,68,78]
admins = [22,13,24,10,16,19]
```

```python
plt.figure(figsize = (12,6))

plt.bar(Platform,Users,ec = 'k',lw = 3 )
plt.bar (Platform,admins,ec = 'k',lw = 3 )

plt.title('Users of apps',fontsize = 20)
plt.xlabel('apps',fontsize = 20)
plt.ylabel('no of users',fontsize = 20)
plt.show()
```

```python
both_indexes = np.arange(len(Platform))
width = 0.45

st.use('ggplot')
plt.figure(figsize = (12,6))

plt.bar(both_indexes,Users,ec = 'k',lw = 3,width = width,label = 'Users' )
plt.bar(both_indexes + width, admins, ec = 'k',lw = 3,width = width ,label = 'admins')

plt.xticks(both_indexes +0.25,Platform,color = 'b',fontsize = 15)
plt.yticks(color = 'b',fontsize = 15)

plt.title('Users & admins of apps ',fontsize = 20)
plt.xlabel('apps',fontsize = 20,c = 'k')
plt.ylabel('Number of users & admins',fontsize = 20,c = 'k')
plt.legend()
plt.show()
```

```python
both_indexes = np.arange(len(Platform))
width = 0.45
#its impossible to  use width = width ,have to use just width

st.use('ggplot')
plt.figure(figsize = (12,6))

plt.barh(both_indexes,Users,width,ec = 'k',lw = 3,label = 'Users' )
plt.barh(both_indexes + width,admins,width, ec = 'k',lw = 3,label = 'admins') 
plt.yticks(both_indexes +0.25,Platform,color = 'b',fontsize = 15)
plt.xticks(color = 'b',fontsize = 15)

plt.title('Users & admins of apps ',fontsize = 20)
plt.ylabel('apps',fontsize = 20,c = 'k')
plt.xlabel('Number of users & admins',fontsize = 20,c = 'k')
plt.grid(c ='gray')
plt.legend()
plt.show()
```
