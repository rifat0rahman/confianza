```python
Classes = ['FIVE','SIX','SEVEN','EIGHT','NINE','TEN']
male_Students = [40,56,45,60,42,50]
female_students= [34,32,43,20,35,25]
plt.figure(figsize=(10,5))
plt.bar(Classes,male_Students,width = 0.6,align='edge',color = 'gray',edgecolor='k',linewidth=1,fill=True)
plt.title('Students of Classes')
plt.ylabel('Class')
plt.xlabel('number of students')
```

```python
classes_index = np.arange(len(Classes))
width=0.4

plt.figure(figsize=(12,5))

plt.bar(classes_index,male_Students,width,align='edge',color = 'gray',edgecolor='k',linewidth=1,
        fill=True,label='male')
plt.bar(classes_index + width,female_students,width,align='edge',color = 'black',edgecolor='k',linewidth=1,
        fill=True,label='female')
plt.xticks(classes_index + width,Classes)
plt.title('Students of Classes')
plt.xlabel('Class')
plt.ylabel('number of students')
plt.legend()
plt.show()
```

```python
classes_index = np.arange(len(Classes))
width=0.4

plt.figure(figsize=(12,5))

plt.barh(classes_index,male_Students,width,align='edge',color = 'gray',edgecolor='k',linewidth=1,
        fill=True,label='male')
plt.barh(classes_index + width,female_students,width,align='edge',color = 'black',edgecolor='k',linewidth=1,
        fill=True,label='female')
plt.yticks(classes_index + width,Classes)
plt.title('Students of Classes')
plt.ylabel('Class')
plt.xlabel('number of students')
plt.legend()
```
